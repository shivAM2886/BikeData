```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
def part_of_day(x):
    if x == 8 or x == 9:
        return "morning"
    elif x == 10 or x == 11:
        return "late_morning"
    elif x == 12 or x == 13 or x == 14:
        return "mid_day"
    elif x == 15 or x == 16 or x == 17 or x ==18:
        return "evening"
    elif x == 19 or x == 20:
        return "late_evening"
```

```python
# initialize dictionary with availability as value and street 
# as key 

def initialize_dict():
    street_array = [i for i in range(1, 8438)]
    dict_avail = {}
    
    for street in street_array:
        dict_avail[street] = 0
    
    return dict_avail
```

```python
def edit_dataframe(x, df):
    if x == 'all':
        return df
    else:
        df = df[df['part_of_day'] == x]
        df = df.reset_index(drop = True)
        return df 
```

```python
avail_type_array = ['all', 'morning', 'late_morning', 'mid_day', 'evening', 'late_evening']
avail_dict_array = []

company_array = ['bird', 'helbiz', 'jump_bikes', 'jump_scooters', 'lime', 'lyft', 'razor', 'skip', 'spin']
week_array = ['week9', 'week10', 'week11', 'week12']
```

```python

for avail in avail_type_array:
    dict_avail = initialize_dict()
    
    for week in week_array:
        print(week)
        
        for company in company_array:
            print(company)
            
            df_file_path = "D:/trip_datasets/dataset_availability/1_attribute_matrix/" + \
            company + "/" + company + "_" + week +".csv"
            
            df = pd.read_csv(df_file_path)
            len1 = len(df)
            
            df['time2'] = pd.to_datetime(df['my_time_stamp2'])
            df['hour_of_day'] = df['time2'].apply(lambda x: x.hour)
            df = df.drop(columns = ['Unnamed: 0'])
            df = df[(df['hour_of_day'] >= 8) & (df['hour_of_day'] <= 20)]
            df['part_of_day'] = df['hour_of_day'].apply(lambda x: part_of_day(x))
            df = edit_dataframe(avail, df)
            len2 = len(df)
            
            print(len1, len2)
            
            unique_time_steps = np.unique(df['my_time_stamp'].values)
            len_unique_time_steps = len(unique_time_steps)
            
            temp_dict = dict(Counter(df['nearest street'].values))
            
            for key, value in temp_dict.items():
                if (key != -1800) and (key != -10):
                    dict_avail[key] = dict_avail[key] + (temp_dict[key]/len_unique_time_steps)
        print("--------------")
        
    avail_dict_array.append(dict_avail)
```

```python
data = {}

for i in range(len(avail_type_array)):
    temp_dict = avail_dict_array[i]
    
    temp_street = np.array(list(temp_dict.keys()))
    temp_array = np.array(list(temp_dict.values()))
    
    street_key_word =  'street_' + avail_type_array[i]
    mean_avail_key_word = 'mean_avail_'+ avail_type_array[i]
    
    data[street_key_word] = temp_street
    data[mean_avail_key_word] = temp_array
    

df_out = pd.DataFrame(data = data)
print(df_out.info())   
```

```python
# week average
avail_day = ['all', 'morning', 'late_morning', 'mid_day', 'evening', 'late_evening']
for avail in avail_day:
    column_name = 'mean_avail_' + avail
    df_out[column_name] = df_out[column_name]/len(week_array)
```

```python
df_out.to_csv("mean_availability_final.csv")
```
